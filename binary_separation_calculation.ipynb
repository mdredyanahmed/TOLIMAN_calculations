```python
import jax
from matplotlib import colormaps
import matplotlib.pyplot as plt
import matplotlib as mpl
import dLuxToliman as dlT
import jax.numpy as np
import optax
from tqdm.notebook import tqdm
import dLux.utils as dlu
import jax.random as jr
import zodiax as zdx

jax.config.update("jax_enable_x64", True)

# Wavefront properties
diameter = 0.125  # Diameter of the aperture in meters
wf_npixels = 256  # Number of pixels across the wavefront
oversample = 4  # Oversampling factor for constructing the aperture

# PSF Parameters
psf_npixels = 128  # Number of pixels across the PSF
psf_pixel_scale = 0.375  # Pixel scale in arcseconds per pixel

# Zernike aberrations
zernike_indexes = np.arange(4, 11)
coeffs = 1e-7 * jr.normal(jr.PRNGKey(0), zernike_indexes.shape)
coords = dlu.pixel_coords(wf_npixels, diameter)
basis = dlu.zernike_basis(zernike_indexes, coords, diameter)

circle = dlu.circle(coords, diameter / 2)
spider = dlu.spider(coords, diameter / 6, [90])
transmission = dlu.combine([circle, spider], oversample)
optical_system = dlT.TolimanOpticalSystem(
    wf_npixels=wf_npixels,
    psf_npixels=psf_npixels,
    oversample=oversample,
    noll_indices=zernike_indexes,
)

param = "aperture.coefficients"
optical_system = optical_system.set(param, 2e-8 * jr.normal(jr.PRNGKey(2), (len(coeffs),)))

optical_system = optical_system.multiply(param, 0)
support = optical_system.aperture.transmission

support_mask = support.at[support < 0.5].set(np.nan)

opd = optical_system.aperture.eval_basis()
optical_system = optical_system.multiply("aperture.basis", 1e-9)
cmap = colormaps['inferno']
cmap.set_bad('k', .5)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(support_mask * opd * 1e6, cmap=cmap)
plt.title("Aberrations")
plt.colorbar(label='um')

#  Parameters
separation = np.array(7.40)
position_angle = np.array(30.0)
x_position = np.array(0.0)
y_position = np.array(0.0)
log_flux = np.array( 7.832)
contrast = np.array(3.203)

# Create a binary star model for the observed systems
star_model = dlT.AlphaCen(
    separation=separation,
    position_angle=position_angle,
    x_position=x_position,
    y_position=y_position,
    log_flux=log_flux,
    contrast=contrast,
)

# Model the star with the Toliman 
model_star = dlT.Toliman(source=star_model, osys=optical_system)
psf_model = model_star.model()

star_data = dlT.AlphaCen(
    separation=separation + 0.1,  # Slightly perturbed
    position_angle=position_angle + 0.1,
    x_position=x_position + 0.001,
    y_position=y_position + 0.001,
    log_flux=log_flux + 0.1,
    contrast=contrast + 0.1,
)

data_star = dlT.Toliman(source=star_data, osys=optical_system)
data_star = data_star.multiply("aperture.coefficients", 0.0)
psf_data = data_star.model()

# Add Gaussian noise
read_noise_std = 5.0

# Define the loss function
@zdx.filter_jit
@zdx.filter_value_and_grad(
    [
        "separation",
        "position_angle",
        "x_position",
        "y_position",
        "log_flux",
        "contrast",
        "aperture.coefficients",
    ]
)
def loss_func(model_star, observed_data, read_noise_std):
    psf_model = model_star.model()
    var = np.maximum(observed_data + read_noise_std**2, read_noise_std**2)
    return -np.nansum(jax.scipy.stats.norm.logpdf(psf_model, loc=observed_data, scale=np.sqrt(var)))


params = [
    "separation",
    "position_angle",
    "x_position",
    "y_position",
    "contrast",
    "log_flux",
    "aperture.coefficients",
]





# Define learning rates for each parameter
learning_rates = {
    "separation": 1e-8,
    "position_angle": 1e-6,
    "x_position": 1e-10,
    "y_position": 1e-10,
    "contrast": 1e-6,
    "log_flux": 1e-9,
    "aperture.coefficients": 1e-15,
}

# Create optimizers for each parameter with their respective learning rates
optimisers = [
    optax.sgd(learning_rates["separation"]),
    optax.sgd(learning_rates["position_angle"]),
    optax.sgd(learning_rates["x_position"]),
    optax.sgd(learning_rates["y_position"]),
    optax.sgd(learning_rates["contrast"]),
    optax.sgd(learning_rates["log_flux"]),
    optax.sgd(learning_rates["aperture.coefficients"]),
]

optim, opt_state = zdx.get_optimiser(model_star, params, optimisers)

# Array to store results
optimized_separations = []
optimized_results = []

# Open a text file to save the results
with open("optimized_separations_1s.txt", "w") as file:
    file.write("Iteration,Optimized_Separation,Position_Angle,X_Position,Y_Position,Contrast,Log_Flux\n")  # Header

    # Generate noisy data and perform optimization for each
    for i in range(100):
        keys = jr.split(jr.PRNGKey(i), 2)
        photons = jr.poisson(keys[0], psf_data)
        read_noise = read_noise_std * jr.normal(keys[1], psf_data.shape)
        observed_data = photons + read_noise

        # Gradient Descent loop
        losses, models_out = [], []
        param_history = {'x_position': [], 'y_position': [],'contrast': [],'log_flux': [],'separation': [],'position_angle': [],'aperture.coefficients': []}
        model_star_loop = model_star
        opt_state_loop = opt_state
        

        with tqdm(range(100), desc=f"Gradient Descent for Iteration {i}") as t:
            for j in t:
                loss, grads = loss_func(model_star_loop, observed_data, read_noise_std)
                updates, opt_state_loop = optim.update(grads, opt_state_loop)
                model_star_loop = zdx.apply_updates(model_star_loop, updates)
                losses.append(loss)
                t.set_description(f"Loss {loss:.5f}")
                # Update the progress bar with the current loss
        
        # Store the parameter history
                param_history['separation'].append(model_star_loop.source.separation)
                param_history['position_angle'].append(model_star_loop.source.position_angle)
                param_history['x_position'].append(model_star_loop.source.x_position)
                param_history['y_position'].append(model_star_loop.source.y_position)
                param_history['contrast'].append(model_star_loop.source.contrast)
                param_history['log_flux'].append(model_star_loop.source.log_flux)
                param_history['aperture.coefficients'].append(model_star_loop.aperture.coefficients)



       
                
                
    
                
                
                
                
                

                optimized_separation = float(model_star_loop.source.separation)
                optimized_results.append(
            [
                optimized_separation,
                float(model_star_loop.source.position_angle),
                float(model_star_loop.source.x_position),
                float(model_star_loop.source.y_position),
                float(model_star_loop.source.contrast),
                float(model_star_loop.source.log_flux),
            ]
        )
        optimized_separations.append(optimized_separation)

        # Save to text file
        file.write(
            f"{i},{optimized_separation},{model_star_loop.source.position_angle},"
            f"{model_star_loop.source.x_position},{model_star_loop.source.y_position},"
            f"{model_star_loop.source.contrast},{model_star_loop.source.log_flux}\n"
        )
# optimized_separation = model_star.source.separation
print("Optimized Separations:", optimized_separations)
print("optimal x_position:", model_star_loop.source.x_position)
print("optimal y_position:", model_star_loop.source.y_position)
print("optimal contrast:", model_star_loop.source.contrast)
print("optimal log_flux:", model_star_loop.source.log_flux)
print("optimal coefficients:", model_star_loop.aperture.coefficients)
import matplotlib.pyplot as plt

# Plot the loss curve
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(losses)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.grid(True)

plt.savefig("loss_curve_1s_frame.png")
# Plot the loss curve for the last 50 epochs
plt.figure(figsize=(12, 6))
plt.plot(range(len(losses) - 50, len(losses)), losses[-50:], marker='o', label="Loss (Last 50 Epochs)")
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss Curve (Last 50 Epochs)')
plt.grid(True)
plt.savefig("loss_curve_last_50_epochs_1s.png")
num_params = len(param_history)
fig, axs = plt.subplots(num_params, 1, figsize=(10, 4 * num_params), sharex=True)

for i, (param_name, values) in enumerate(param_history.items()):
    axs[i].plot(values, label=f"{param_name}")
    axs[i].set_ylabel("Parameter Value")
    axs[i].set_title(f"{param_name} ")
    axs[i].legend()

plt.xlabel("Iteration")
plt.tight_layout()
plt.savefig("loss_curve_params_1s_frame.png")
plt.show()








# recovered_psf_model = model_star.model()
recovered_psf_data = data_star.model()

# Update the residual plot to use the 'seismic' colormap
cmap_residual = colormaps['seismic']
cmap_residual.set_bad('k', .5)

fig, axs = plt.subplots(2, 3, figsize=(18, 12))

# Plot the model PSF
im0 = axs[0, 0].imshow(observed_data, cmap=cmap)
axs[0, 0].set_title("Observed data")
axs[0, 0].set_xlabel("X Pixels")
axs[0, 0].set_ylabel("Y Pixels")
cbar0 = plt.colorbar(im0, ax=axs[0, 0])
cbar0.set_label('Photons')

# Plot the data PSF
im1 = axs[0, 1].imshow(recovered_psf_data, cmap=cmap)
axs[0, 1].set_title("Recovered Data")
axs[0, 1].set_xlabel("X Pixels")
axs[0, 1].set_ylabel("Y Pixels")
cbar1 = plt.colorbar(im1, ax=axs[0, 1])
cbar1.set_label('Photons')

# Plot the residuals with the 'seismic' colormap
residuals = recovered_psf_data - observed_data
vmin, vmax = -np.max(np.abs(residuals)), np.max(np.abs(residuals))
im2 = axs[0, 2].imshow(residuals, cmap=cmap_residual, vmin=vmin, vmax=vmax)

axs[0, 2].set_title("Residual")
axs[0, 2].set_xlabel("X Pixels")
axs[0, 2].set_ylabel("Y Pixels")
cbar2 = plt.colorbar(im2, ax=axs[0, 2])
cbar2.set_label('Photons')

# Plot the histogram of residuals
axs[1, 0].hist(residuals.ravel(), bins=50, color='blue', alpha=0.7)
axs[1, 0].set_title("Histogram of Residuals")
axs[1, 0].set_xlabel("Residual Value")
axs[1, 0].set_ylabel("Frequency")

# Plot the z-score of residuals
z_score = (residuals) / np.sqrt(np.maximum(observed_data + read_noise_std**2, read_noise_std**2))
# Plot the histogram of z-scores with a unit normal curve
z_score_flat = z_score.ravel()
axs[1, 2].hist(z_score_flat, bins=50, density=True, alpha=0.6, color='g', label='Z-Score Histogram')

# Plot the unit normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)
axs[1, 2].plot(x, p, 'k', linewidth=2, label='Unit Normal Distribution')

axs[1, 2].set_title("Histogram of Z-Scores with Unit Normal")
axs[1, 2].set_xlabel("Z-Score")
axs[1, 2].set_ylabel("Density")
axs[1, 2].legend()
vmin, vmax = -np.max(np.abs(z_score)), np.max(np.abs(z_score))
im3 = axs[1, 1].imshow(z_score, cmap=cmap_residual, vmin=vmin, vmax=vmax)

axs[1, 1].set_title("Z-Score of Residuals")
axs[1, 1].set_xlabel("X Pixels")
axs[1, 1].set_ylabel("Y Pixels")
cbar3 = plt.colorbar(im3, ax=axs[1, 1])
cbar3.set_label('Z-Score')

# Hide the empty subplot
axs[1, 2].axis('off')

plt.tight_layout()
plt.savefig("model_comparison_1s_frame.png", dpi=400)
plt.show()
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Read the data from the file
data = np.loadtxt("optimized_separations_10000s.txt", delimiter=",", skiprows=1)

# Extract the params
optimized_separations = data[:, 1]

position_angle = data[:, 2]
x_position= data[:, 3]
Y_position= data[:, 4]
contrast= data[:, 5]
log_flux= data[:, 6]
print(f'Mean of Optimized Separations: {np.mean(optimized_separations)}')
print(f'Mean of Position Angle: {np.mean(position_angle)}')
print(f'Mean of X Position: {np.mean(x_position)}')
print(f'Mean of Y Position: {np.mean(Y_position)}')
print(f'Mean of Contrast: {np.mean(contrast)}')
print(f'Mean of Log Flux: {np.mean(log_flux)}')


# Perform a statistical test (e.g., Shapiro-Wilk test for normality)
stat, p_value = stats.shapiro(optimized_separations)
print(f'Shapiro-Wilk Test: Statistics={stat}, p-value={p_value}')

# Fit a Gaussian to the histogram data
mu, std = stats.norm.fit(optimized_separations)

# Plot the histogram with the Gaussian fit
plt.figure(figsize=(10, 6))
plt.hist(optimized_separations, bins=20, color='blue', alpha=0.7, label='Data')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std) * len(optimized_separations) * (xmax - xmin) / 20  # Scale the PDF to match the histogram
plt.plot(x, p, 'k', linewidth=2, label=f'Gaussian fit: $\mu={mu:.7f}$, $\sigma={std:.7f}$')
plt.xlabel('Optimized Separation')
plt.ylabel('Frequency')
plt.title('Histogram of Optimized Separations with Gaussian Fit')
plt.legend()
plt.grid(True)
# plt.savefig("histogram_optimized_separations_10000s_frame.png", dpi=400)
plt.show()


```
